```python
###################################################################
# FILE 1
# QUERY TIME
import matplotlib.pyplot as plt
import numpy as np

pattern1_file1_values = [131, 81.1, 62.6, 133]
pattern2_file1_values = [187, 126, 18, 28.4]
pattern3_file1_values = [110, 84.8, 0.472, 0.506]

algorithms = ['Sorted Index', 'Hash Table', 'Suffix Array', 'Suffix Tree']

_X = np.arange(len(algorithms))

plt.figure(1, figsize=(9,5))

plt.bar(_X-0.2, pattern1_file1_values, 0.2, label='ATGATG')
plt.bar(_X, pattern2_file1_values, 0.2, label='CTCTCTA')
plt.bar(_X+0.2, pattern3_file1_values, 0.2, label='TCACTACTCTCA')

plt.xticks(_X, algorithms) # set labels manually

plt.legend(loc=1, prop={'size': 10})
plt.xlabel("Algorithms")
plt.ylabel("query time [ms]")

plt.suptitle('Query Time - Canis lupus familiaris genome, chromosome 1')

plt.show()
```
